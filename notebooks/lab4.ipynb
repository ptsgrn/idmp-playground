```python
from owlready2 import get_ontology

# Load IDMP ontology
onto = get_ontology(
    "file://data/ISO11615-MedicinalProducts-Merged.rdf").load()

# Print the ontology base URI
print(f"Ontology loaded: {onto.base_iri}")
```

```python
for cls in onto.classes():
    print(cls)
    
print(onto.name)
```

```python
onto
```

```python
# Retrieve all medicinal products and their substances

for product in onto.MedicinalProduct.instances():
    print(f"Product: {product.name}")
    for substance in product.hasActiveIngredient:
        print(f"  - Contains: {substance.name}")
```
